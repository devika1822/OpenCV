```python
import cv2
```

```python
image=cv2.imread('Pictures/Saved Pictures/colours.jpeg')
image_1= cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
green_lo=np.array([0,10,0])
green_hi=np.array([255,200,255])
mask=cv2.inRange(image,green_lo,green_hi)
image[mask>0]=(255,0,0)
cv2.imwrite('result.jpg',image)
```
